```python
from sklearn.datasets import load_iris
from pymfe.mfe import MFE
```

# Grupos de Meta-features

```python
model = MFE()
model_groups = model.valid_groups()
print(model_groups)
```

| Grupo de meta-features | ¿Qué información capturan? | ¿Para qué sirven? | ¿Cuándo se usan? | ¿Cómo se interpretan? | Tipo de variables |
|------------------------|----------------------------|-------------------|------------------|-----------------------|-------------------|
| **General** | Propiedades estructurales básicas del dataset (número de instancias, atributos, clases, dimensionalidad). | Describir tamaño, forma y composición básica del dataset. | Siempre como línea base en cualquier problema. | No reflejan complejidad estadística; indican restricciones estructurales. | Numéricas y categóricas. |
| **Statistical** | Distribuciones estadísticas de atributos numéricos (media, varianza, correlación, asimetría, curtosis, sparsity). | Caracterizar la estructura estadística interna del dataset. | Meta-Learning en clasificación y regresión. | Correlación, dispersión y asimetría indican necesidad de transformaciones o modelos específicos. | Principalmente numéricas (categóricas transformadas). |
| **Information Theory** | Medidas de información y dependencia (entropía, información mutua). | Cuantificar relevancia de atributos y predictibilidad del target. | Problemas no lineales y con variables categóricas. | Alta entropía implica incertidumbre; alta información mutua indica atributos relevantes. | Categóricas y numéricas (mediante discretización). |
| **Model-based** | Propiedades estructurales de un modelo inducido (p. ej. profundidad y tamaño de un árbol). | Aproximar la complejidad del problema desde un modelo concreto. | Principalmente en clasificación. | Árboles grandes indican relaciones complejas. | Principalmente numéricas, dependientes del modelo. |
| **Landmarking** | Desempeño de algoritmos simples y rápidos. | Estimar rápidamente la dificultad del problema y familias de modelos prometedoras. | Clasificación y AutoML con restricciones de tiempo. | Buen desempeño implica problema sencillo. | Numéricas, con target discreto. |
| **Relative Landmarking** | Rangos relativos de desempeño entre algoritmos simples. | Comparar algoritmos evitando dependencia de métricas específicas. | Meta-Learning comparativo. | Rangos bajos indican mejor desempeño relativo. | Numéricas, clasificación. |
| **Subsampling Landmarking** | Desempeño de algoritmos sobre submuestras del dataset. | Reducir costo computacional y evaluar estabilidad. | Datasets grandes y AutoML. | Alta variabilidad indica inestabilidad del desempeño. | Numéricas, clasificación. |
| **Clustering** | Estructura de agrupamiento natural (cohesión y separación). | Evaluar separabilidad y estructura interna sin usar el target. | Aprendizaje no supervisado o previo a clasificación. | Alta separación indica grupos bien definidos. | Principalmente numéricas; sensibles a escala y NaN. |
| **Concept** | Variabilidad y cohesión intra-clase. | Evaluar consistencia del concepto de clase. | Clasificación. | Alta variabilidad indica clases poco definidas. | Numéricas. |
| **Itemset** | Co-ocurrencias y correlaciones entre atributos binarios. | Identificar patrones frecuentes. | Datos transaccionales y discretos. | Alta frecuencia indica patrones comunes. | Binarias o categóricas discretizadas. |
| **Complexity** | Dificultad geométrica, densidad y solapamiento entre clases. | Estimar dificultad intrínseca del problema. | Clasificación y análisis profundo del dataset. | Valores altos indican problemas complejos. | Numéricas; sensibles a ruido y valores faltantes. |

```python
data = load_iris()
y = data.target
X = data.data

mfe = MFE(groups=["complexity"])
mfe.fit(X, y)
ft = mfe.extract()
print("\n".join("{:50} {:30}".format(x, y) for x, y in zip(ft[0], ft[1])))
```
